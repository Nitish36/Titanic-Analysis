```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("datasets/titanic.csv")
dataset.head(20)
```

```python
dataset.info()
```

### Data Preprocessing

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
new_dataset = dataset.fillna(0)
new_dataset.isnull().sum()
```

### Data Analysis

```python
#To find the count of survivors
new_dataset['survived'].value_counts()
```

```python
survived_counts = new_dataset['survived'].value_counts()

# Plotting the bar graph
plt.rcParams['figure.figsize'] = [6, 5]  # Set figure size
plt.bar(survived_counts.index, survived_counts.values, color=['blue','green'])
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Count')
plt.xticks([0,1])
# Adding labels to the bars
for i, count in enumerate(survived_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=10)

# Displaying the bar graph
plt.show()
```

```python
#To find the count of pclasses
new_dataset['pclass'].value_counts()
```

```python
pclass_counts = new_dataset['pclass'].value_counts()

# Plotting the pie chart
plt.rcParams['figure.figsize'] = [6, 6]  # Set figure size
plt.pie(pclass_counts.values, labels=pclass_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'red'])
plt.title('Passenger Class Distribution')

# Displaying the pie chart
plt.show()
```

```python
#To find the count of males and females present
new_dataset['sex'].value_counts()
```

```python
gender_count = new_dataset['sex'].value_counts()

# Plotting the bar graph
plt.rcParams['figure.figsize'] = [6, 4]  # Set figure size
plt.bar(gender_count.index, gender_count.values, color=['blue','purple'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Count')
plt.xticks([0,1])
# Adding labels to the bars
for i, count in enumerate(gender_count.values):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=10)

# Displaying the bar graph
plt.show()
```

```python
alive_count = new_dataset['alive'].value_counts()

# Plotting the bar graph
plt.rcParams['figure.figsize'] = [6, 6]  # Set figure size
plt.bar(alive_count.index, alive_count.values, color=['salmon', 'lightgreen'])
plt.xlabel('Alive')
plt.ylabel('Count')
plt.title('Alive Count')

# Adding labels to the bars
for i, count in enumerate(alive_count.values):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=10)

# Set x-axis ticks and labels
plt.xticks(alive_count.index, ['No', 'Yes'])

# Adding grid lines
#plt.grid(axis='y', linestyle='--', alpha=0.7)

# Displaying the bar graph
plt.show()
```

```python
ages = new_dataset['age']

# Plotting the histogram
plt.rcParams['figure.figsize'] = [8, 6]  # Set figure size
plt.hist(ages, bins=20, color='skyblue', edgecolor='black')  # You can adjust the number of bins for different granularity
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

# Displaying the histogram
plt.show()
```

```python
new_datasetDF = pd.DataFrame(new_dataset)
new_datasetDF
```

```python
#To find how gender is related with class
genderclass = new_datasetDF.groupby('class')['sex'].value_counts()
genderclassDF = pd.DataFrame(genderclass)
genderclassDF
```

```python
genderclass = new_datasetDF.groupby('class')['sex'].value_counts()

# Creating a DataFrame from the grouped data
genderclassDF = pd.DataFrame(genderclass)

# Unstacking the DataFrame to pivot 'sex' as columns
genderclassDF = genderclassDF.unstack()

# Plotting the bar graph
plt.rcParams['figure.figsize'] = [8, 6]  # Set figure size
genderclassDF.plot.bar(color=['skyblue', 'lightcoral'], edgecolor='black')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Gender Distribution in Each Passenger Class')

# Displaying the bar graph
plt.show()
```

```python
#To find out how many were alone
whoalone = new_datasetDF.groupby('who')['alone'].value_counts()
whoaloneDF = pd.DataFrame(whoalone)
whoaloneDF
```

```python
whoaloneclass = new_datasetDF.groupby('who')['alone'].value_counts()

# Creating a DataFrame from the grouped data
whoaloneDF = pd.DataFrame(whoalone)

# Unstacking the DataFrame to pivot 'sex' as columns
whoaloneDF = whoaloneDF.unstack()

# Plotting the bar graph
plt.rcParams['figure.figsize'] = [8, 6]  # Set figure size
whoaloneDF.plot.bar(color=['skyblue', 'lightcoral'], edgecolor='black')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Alone Distribution of Each Passenger ')

# Displaying the bar graph
plt.show()
```

```python
#Count of each person and where they embarked to
whoembarked = new_datasetDF.groupby('who')['embark_town'].value_counts()
whoembarkedDF = pd.DataFrame(whoembarked)
whoembarkedDF
```

```python
# Grouping by 'who' and 'embark_town' and getting the value counts
whoembarked = new_datasetDF.groupby('who')['embark_town'].value_counts()

# Creating a DataFrame from the grouped data
whoembarkedDF = pd.DataFrame(whoembarked)

# Unstacking the DataFrame to pivot 'embark_town' as columns
whoembarkedDF = whoembarkedDF.unstack()

# Plotting the bar graph
plt.rcParams['figure.figsize'] = [8, 6]  # Set figure size
whoembarkedDF.plot.bar(color=['skyblue', 'lightcoral', 'lightgreen'], edgecolor='black')
plt.xlabel('Person')
plt.ylabel('Count')
plt.title('Count of Each Person and Where They Embarked To')

# Displaying the bar graph
plt.show()
```
